```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data2/all_states_all_years_reverse_coded_v1.csv')
data = pd.DataFrame(data)

data['Location Returned Split'] = data['Location Returned'].str.split(", ", expand = False)
data.head()
```

```python
temp = data['Location Returned Split']
temp.astype(str)

print(len(temp))
# testing
temp[0][-2]
```

```python
new = [None]*len(temp)
for i in range(0, len(temp)):
    new[i] = temp[i][-2]
```

```python
new = pd.DataFrame(new, columns = ['PostalCode'])
new.head()
```

```python
print(new['PostalCode'][2].isnumeric())
print(new['PostalCode'][0].isnumeric())
```

```python
data['Postal Code'] = new
data.head()
```

```python
data = data[data['Postal Code'].apply(lambda x: x.isnumeric())]
data.head()
```

```python
# export csv
data.to_csv('data2/Postal Code.csv', index=False)
```
