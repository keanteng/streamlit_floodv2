# Run With Folium Using Kaggle Example

```python
import streamlit as st
import folium
from folium.plugins import HeatMap
import pandas as pd

# Create a base map
m_5 = folium.Map(location=[5.32,100.0589], tiles='cartodbpositron', zoom_start=12)

test = pd.read_csv("Flood Data Updated Geocoded.csv")

# Add a heatmap to the base map
HeatMap(data=test[['Latitude', 'Longitude']], radius=10).add_to(m_5)

# Display the map
m_5
```
